**Importing libraries(like pandas) for data manipulation**

```python
import pandas as pd
```

____
**Generating data for our regression model**  

    predictors(X) = no of time radio ads shown on television
    target(Y) = sales of radio

```python
df = pd.DataFrame(list(zip(list(range(20)), list(range(0,40,2)))), columns =['radio_ads', 'sales'])
df.head()
```

___
**Defining a function for predicting sales of radio**

```python
def prediction(weight, radio_ads, bais):
    return weight * radio_ads + bais
```

___
**Defining Cost Function**  
    
this function will calculate MSE(Mean Square Error)

```python
def cost_function(radio_ads, sales, weight, bais):
    N = len(radio_ads)
    total_error = 0.0
    for i in range(N):
        total_error += (sales[i] - (weight * radio_ads[i] + bais))**2
    return total_error / N
```

___
**Define Dredient Decent function**

this function will help in updating weights and bias so that error gets minimized  
Learning rate = steps size taken for weights and bias optimization  
            
           High learning rate = overshooting
           Low learning rate = 
        
        

```python
def update_weights(radio_ads, sales, weight, bais, learning_rate):
    weight_derive = 0.0
    bais_derive = 0.0
    N = len(radio_ads)
    for i in range(N):
        weight_derive += -2*radio_ads[i]*(sales[i] - (weight * radio_ads[i] + bais))
        bais_derive += -2*(sales[i] - (weight * radio_ads[i] + bais))
    weight -= (weight_derive/N)*learning_rate
    bais -= (bais_derive/N)*learning_rate
    return weight, bais
```

___
**Training function**  
This function will help in training of the model and will get best weight and bias value for prediction

```python
def train(radio_ads, sales, weight, bais, learning_rate, iters):
    cost_history = []
    for i in range(iters):
        weight, bais = update_weights(radio_ads, sales, weight, bais, learning_rate)
        cost = cost_function(radio_ads, sales, weight, bais)
        cost_history.append(cost)
        if i % 10 == 0:
            print("iter = {:d}    weight = {:.2f}    bais = {:.4f}    cost = {:.2}".format(i, weight, bais, cost))
    return weight, bais, cost_history
```

```python
weight, bais, cost_history = train(df["radio_ads"], df["sales"], 10, 10, 0.001, 10000)
```

```python
print(weight)
print(bais)
```

___
Prediction of sales for radio_ads = 20

```python
print(prediction(weight, 20, bais))
```
